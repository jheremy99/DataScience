```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Calcular el ingreso total (suma de la columna Precio) para cada tienda
ingreso_tienda = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()

# Crear una direccciòn de cada ingreso
ingresos = {
    'Tienda 1': ingreso_tienda,
    'Tienda 2': ingreso_tienda2,
    'Tienda 3': ingreso_tienda3,
    'Tienda 4': ingreso_tienda4
}

# Mostrara los ingresos totales de cada tienda
print("Ingresos totales por tienda:")
for tienda, ingreso in ingresos.items():
    print(f"{tienda}: ${ingreso:,.2f}")


# Visualizar los resultados con un gráfico de barras mejor comprenciòn
plt.figure(figsize=(10, 6))
plt.bar(list(ingresos.keys()), list(ingresos.values()), color='skyblue')
plt.title('Ingreso Total por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Ingreso Total ($)')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Añadira etiquetas de valor en las barras
for i, v in enumerate(ingresos.values()):
    plt.text(i, v + max(ingresos.values())*0.01, f'${v:,.2f}', ha='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"


tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Añadir una columna para identificar cada tienda
tienda1['Tienda'] = 'Tienda 1'
tienda2['Tienda'] = 'Tienda 2'
tienda3['Tienda'] = 'Tienda 3'
tienda4['Tienda'] = 'Tienda 4'

# Combinar todos los datos en un solo DataFrame para facilitar el análisis
todas_tiendas = pd.concat([tienda1, tienda2, tienda3, tienda4])


# Análisis por tienda
resultados_por_tienda = {}
tiendas_df = [tienda1, tienda2, tienda3, tienda4]
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Función para obtener las categorías más populares
def obtener_categorias_populares(df, nombre_tienda):
    # Verificar que el DataFrame contiene la columna de categoría
    categoria_col = 'Categoría del Producto'

    if categoria_col not in df.columns:
        # Si la columna no existe exactamente, buscar otras opciones
        posibles_columnas = [col for col in df.columns if 'categ' in col.lower()]
        if posibles_columnas:
            categoria_col = posibles_columnas[0]
        else:
            print(f"No se encontró una columna de categoría en {nombre_tienda}")
            return None

    ventas_por_categoria = df[categoria_col].value_counts().reset_index()
    ventas_por_categoria.columns = ['Categoría', 'Cantidad']
    return ventas_por_categoria

for i, tienda_df in enumerate(tiendas_df):
    nombre_tienda = nombres_tiendas[i]
    resultados = obtener_categorias_populares(tienda_df, nombre_tienda)
    if resultados is not None:
        resultados_por_tienda[nombre_tienda] = resultados

for nombre_tienda, resultados in resultados_por_tienda.items():
    print(f"\n{nombre_tienda} - Categorías más populares:")
    print(resultados.head())


plt.figure(figsize=(16, 12))

# Crear un gráfico para cada tienda usando matplotlib directamente
for i, (nombre_tienda, resultados) in enumerate(resultados_por_tienda.items(), 1):
    plt.subplot(2, 2, i)

    # Limitar a las 5 categorías más populares para mejor visualización
    top_categorias = resultados.head(5)

    # Colores para las barras
    colors = plt.cm.viridis(np.linspace(0, 0.9, len(top_categorias)))

    # Crear gráfico de barras con matplotlib
    bars = plt.barh(top_categorias['Categoría'], top_categorias['Cantidad'], color=colors)
    plt.title(f'{nombre_tienda} - Top 5 Categorías más Populares')
    plt.xlabel('Cantidad Vendida')
    plt.ylabel('Categoría')

    # Añadir valores a las barras
    for bar in bars:
        width = bar.get_width()
        plt.text(width + 0.3, bar.get_y() + bar.get_height()/2,
                f'{int(width)}', ha='left', va='center')

plt.tight_layout()
plt.show()

# Análisis global (todas las tiendas juntas)
print("\nAnálisis global de todas las tiendas:")
ventas_globales = obtener_categorias_populares(todas_tiendas, "Todas las tiendas")
print(ventas_globales.head(10))

# Gráfico de las categorías más populares en general - VERSIÓN CORREGIDA
plt.figure(figsize=(12, 6))
top_categorias_global = ventas_globales.head(10)

# Colores para las barras
colors = plt.cm.coolwarm(np.linspace(0, 0.9, len(top_categorias_global)))

# Crear gráfico con matplotlib
bars = plt.barh(top_categorias_global['Categoría'], top_categorias_global['Cantidad'], color=colors)
plt.title('Top 10 Categorías más Populares en Todas las Tiendas')
plt.xlabel('Cantidad Total Vendida')
plt.ylabel('Categoría')

# Añadir valores a las barras
for bar in bars:
    width = bar.get_width()
    plt.text(width + 0.5, bar.get_y() + bar.get_height()/2,
            f'{int(width)}', ha='left', va='center')

plt.tight_layout()
plt.show()


top5_categorias = ventas_globales.head(5)['Categoría'].tolist()

# Crear un DataFrame para la comparación entre tiendas
comparacion_tiendas = pd.DataFrame()

for nombre_tienda, tienda_df in zip(nombres_tiendas, tiendas_df):
    # Obtener la columna de categoría
    categoria_col = 'Categoría del Producto'
    if categoria_col not in tienda_df.columns:
        posibles_columnas = [col for col in tienda_df.columns if 'categ' in col.lower()]
        if posibles_columnas:
            categoria_col = posibles_columnas[0]

    # Obtener conteo para las top 5 categorías
    for categoria in top5_categorias:
        count = len(tienda_df[tienda_df[categoria_col] == categoria])
        comparacion_tiendas.loc[categoria, nombre_tienda] = count

# Visualizar la comparación
comparacion_tiendas.plot(kind='bar', figsize=(12, 8))
plt.title('Comparación de las 5 Categorías más Populares entre Tiendas')
plt.xlabel('Categoría')
plt.ylabel('Cantidad Vendida')
plt.legend(title='Tienda')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# 3. Calificación promedio de la tienda


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"


tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)


tienda1['Tienda'] = 'Tienda 1'
tienda2['Tienda'] = 'Tienda 2'
tienda3['Tienda'] = 'Tienda 3'
tienda4['Tienda'] = 'Tienda 4'

# Combinar todos los datos en un solo DataFrame para facilitar el análisis
todas_tiendas = pd.concat([tienda1, tienda2, tienda3, tienda4])


# Identificar la columna de calificación
calificacion_col = 'Calificación'
if calificacion_col not in tienda1.columns:
    posibles_columnas = [col for col in tienda1.columns if 'califi' in col.lower() or 'rating' in col.lower()]
    if posibles_columnas:
        calificacion_col = posibles_columnas[0]
    else:
        print("No se encontró una columna de calificación")

# Análisis de calificaciones por tienda
tiendas_df = [tienda1, tienda2, tienda3, tienda4]
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Función para analizar las calificaciones
def analizar_calificaciones(df, nombre_tienda):
    if calificacion_col not in df.columns:
        print(f"La columna {calificacion_col} no existe en {nombre_tienda}")
        return None

    # Obtener estadísticas descriptivas
    calificacion_media = df[calificacion_col].mean()
    calificacion_mediana = df[calificacion_col].median()
    calificacion_max = df[calificacion_col].max()
    calificacion_min = df[calificacion_col].min()
    calificacion_std = df[calificacion_col].std()

    # Contar frecuencia de cada calificación
    distribucion = df[calificacion_col].value_counts().sort_index()

    return {
        'media': calificacion_media,
        'mediana': calificacion_mediana,
        'max': calificacion_max,
        'min': calificacion_min,
        'std': calificacion_std,
        'distribucion': distribucion
    }

# Recopilar resultados por tienda
resultados_calificaciones = {}
for nombre_tienda, df in zip(nombres_tiendas, tiendas_df):
    resultados = analizar_calificaciones(df, nombre_tienda)
    if resultados:
        resultados_calificaciones[nombre_tienda] = resultados

# Mostrar resultados
print("\n=== RESULTADOS DE CALIFICACIONES POR TIENDA ===")
for nombre_tienda, resultados in resultados_calificaciones.items():
    print(f"\n{nombre_tienda}:")
    print(f"  Calificación Promedio: {resultados['media']:.2f}")
    print(f"  Calificación Mediana: {resultados['mediana']:.2f}")
    print(f"  Calificación Mínima: {resultados['min']}")
    print(f"  Calificación Máxima: {resultados['max']}")
    print(f"  Desviación Estándar: {resultados['std']:.2f}")
    print("  Distribución de Calificaciones:")
    print(resultados['distribucion'])

# Crear un DataFrame para visualización comparativa
calificaciones_comparativas = pd.DataFrame({
    'Tienda': nombres_tiendas,
    'Calificación Promedio': [resultados_calificaciones[t]['media'] for t in nombres_tiendas],
    'Desviación Estándar': [resultados_calificaciones[t]['std'] for t in nombres_tiendas]
})

print("\nComparación de Calificaciones Promedio:")
print(calificaciones_comparativas)

# Visualizaciones
# 1. Gráfico de barras para calificaciones promedio
plt.figure(figsize=(12, 6))
bars = plt.bar(calificaciones_comparativas['Tienda'],
               calificaciones_comparativas['Calificación Promedio'],
               yerr=calificaciones_comparativas['Desviación Estándar'],
               capsize=10, color='skyblue', edgecolor='navy', alpha=0.7)

# Añadir etiquetas con los valores
for i, bar in enumerate(bars):
    valor = calificaciones_comparativas['Calificación Promedio'].iloc[i]
    plt.text(bar.get_x() + bar.get_width()/2,
             valor + 0.1,
             f'{valor:.2f}',
             ha='center', fontweight='bold')

plt.title('Calificación Promedio por Tienda', fontsize=16)
plt.xlabel('Tienda', fontsize=14)
plt.ylabel('Calificación Promedio', fontsize=14)
plt.ylim(0, 5.5)  # Ajustar para mostrar barras completas y etiquetas
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# 2. Gráfico de distribución de calificaciones para cada tienda
plt.figure(figsize=(14, 10))

for i, nombre_tienda in enumerate(nombres_tiendas, 1):
    plt.subplot(2, 2, i)

    # Obtener la distribución de calificaciones
    distribucion = resultados_calificaciones[nombre_tienda]['distribucion']

    # Crear gráfico de barras
    bars = plt.bar(distribucion.index, distribucion.values, color='lightgreen', edgecolor='darkgreen', alpha=0.7)

    # Añadir etiquetas con los valores
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2,
                 height + 0.5,
                 f'{int(height)}',
                 ha='center')

    plt.title(f'Distribución de Calificaciones - {nombre_tienda}', fontsize=12)
    plt.xlabel('Calificación', fontsize=10)
    plt.ylabel('Número de Productos', fontsize=10)
    plt.xticks(distribucion.index)
    plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()

# 3. Gráfico de calificaciones por categoría
# Identificar la columna de categoría
categoria_col = 'Categoría del Producto'
if categoria_col not in tienda1.columns:
    posibles_columnas = [col for col in tienda1.columns if 'categ' in col.lower()]
    if posibles_columnas:
        categoria_col = posibles_columnas[0]

# Análisis de calificaciones por categoría (global)
if categoria_col in todas_tiendas.columns:
    # Obtener las top 5 categorías más comunes
    top_categorias = todas_tiendas[categoria_col].value_counts().head(5).index.tolist()

    # Filtrar el DataFrame para incluir solo esas categorías
    df_top_categorias = todas_tiendas[todas_tiendas[categoria_col].isin(top_categorias)]

    # Calcular la calificación media por categoría y tienda
    calificacion_por_categoria = df_top_categorias.groupby([categoria_col, 'Tienda'])[calificacion_col].mean().reset_index()

    # Visualizar resultados
    plt.figure(figsize=(14, 8))

    # Crear gráfico de barras agrupadas
    sns.barplot(x=categoria_col, y=calificacion_col, hue='Tienda', data=calificacion_por_categoria)

    plt.title('Calificación Promedio por Categoría y Tienda', fontsize=16)
    plt.xlabel('Categoría', fontsize=14)
    plt.ylabel('Calificación Promedio', fontsize=14)
    plt.xticks(rotation=45, ha='right')
    plt.ylim(0, 5.5)  # Ajustar para mostrar barras completas
    plt.legend(title='Tienda')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
```

# 4. Productos más y menos vendidos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"


tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)


tienda1['Tienda'] = 'Tienda 1'
tienda2['Tienda'] = 'Tienda 2'
tienda3['Tienda'] = 'Tienda 3'
tienda4['Tienda'] = 'Tienda 4'


todas_tiendas = pd.concat([tienda1, tienda2, tienda3, tienda4])

# Identificar la columna de productos
producto_col = 'Producto'
if producto_col not in tienda1.columns:
    posibles_columnas = [col for col in tienda1.columns if 'product' in col.lower() or 'produc' in col.lower()]
    if posibles_columnas:
        producto_col = posibles_columnas[0]
    else:
        print("No se encontró una columna de productos")

print(f"Usando la columna de productos: {producto_col}")

# Lista de tiendas y sus DataFrames
tiendas_df = [tienda1, tienda2, tienda3, tienda4]
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Función para analizar productos vendidos
def analizar_productos(df, nombre_tienda):
    if producto_col not in df.columns:
        print(f"La columna {producto_col} no existe en {nombre_tienda}")
        return None

    # Contar ventas por producto
    ventas_por_producto = df[producto_col].value_counts().reset_index()
    ventas_por_producto.columns = ['Producto', 'Cantidad']

    # Obtener los 5 productos más vendidos
    top_productos = ventas_por_producto.head(5)

    # Obtener los 5 productos menos vendidos
    bottom_productos = ventas_por_producto.tail(5)

    return {
        'top_productos': top_productos,
        'bottom_productos': bottom_productos,
        'todos_productos': ventas_por_producto
    }

# Recopilar resultados por tienda
resultados_productos = {}
for nombre_tienda, df in zip(nombres_tiendas, tiendas_df):
    resultados = analizar_productos(df, nombre_tienda)
    if resultados:
        resultados_productos[nombre_tienda] = resultados

# Mostrar resultados
print("\n=== ANÁLISIS DE PRODUCTOS POR TIENDA ===")
for nombre_tienda, resultados in resultados_productos.items():
    print(f"\n{nombre_tienda}:")

    print("\nProductos más vendidos:")
    print(resultados['top_productos'])

    print("\nProductos menos vendidos:")
    print(resultados['bottom_productos'])

# Visualizaciones
# 1. Gráficos de productos más vendidos por tienda
plt.figure(figsize=(18, 14))

for i, nombre_tienda in enumerate(nombres_tiendas, 1):
    plt.subplot(2, 2, i)

    top_productos = resultados_productos[nombre_tienda]['top_productos']

    bars = plt.barh(top_productos['Producto'], top_productos['Cantidad'],
                    color='lightgreen', edgecolor='darkgreen', alpha=0.7)

    for bar in bars:
        width = bar.get_width()
        plt.text(width + 0.3, bar.get_y() + bar.get_height()/2,
                 f'{int(width)}', va='center')

    plt.title(f'Top 5 Productos Más Vendidos - {nombre_tienda}', fontsize=14)
    plt.xlabel('Cantidad Vendida', fontsize=12)
    plt.ylabel('Producto', fontsize=12)
    plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()

# 2. Gráficos de productos menos vendidos por tienda
plt.figure(figsize=(18, 14))

for i, nombre_tienda in enumerate(nombres_tiendas, 1):
    plt.subplot(2, 2, i)

    bottom_productos = resultados_productos[nombre_tienda]['bottom_productos']

    bars = plt.barh(bottom_productos['Producto'], bottom_productos['Cantidad'],
                    color='salmon', edgecolor='darkred', alpha=0.7)

    for bar in bars:
        width = bar.get_width()
        plt.text(width + 0.3, bar.get_y() + bar.get_height()/2,
                 f'{int(width)}', va='center')

    plt.title(f'Top 5 Productos Menos Vendidos - {nombre_tienda}', fontsize=14)
    plt.xlabel('Cantidad Vendida', fontsize=12)
    plt.ylabel('Producto', fontsize=12)
    plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()

# 3. Análisis combinado: ¿Los productos más vendidos son los mejor calificados?
calificacion_col = 'Calificación'
if calificacion_col in tienda1.columns:

    plt.figure(figsize=(14, 10))

    for i, nombre_tienda in enumerate(nombres_tiendas, 1):
        plt.subplot(2, 2, i)

        df = tiendas_df[i-1]

        producto_analisis = df.groupby(producto_col).agg({
            'Tienda': 'count',
            calificacion_col: 'mean'
        }).reset_index()

        producto_analisis.columns = ['Producto', 'Cantidad_Vendida', 'Calificación_Promedio']

        scatter = plt.scatter(producto_analisis['Cantidad_Vendida'],
                             producto_analisis['Calificación_Promedio'],
                             alpha=0.6, c=producto_analisis['Cantidad_Vendida'],
                             cmap='viridis', s=100)
        top_productos = producto_analisis.nlargest(3, 'Cantidad_Vendida')
        for _, row in top_productos.iterrows():
            plt.annotate(row['Producto'],
                         (row['Cantidad_Vendida'], row['Calificación_Promedio']),
                         xytext=(5, 5), textcoords='offset points',
                         fontsize=8, fontweight='bold')

        plt.title(f'Relación entre Ventas y Calificaciones - {nombre_tienda}', fontsize=12)
        plt.xlabel('Cantidad Vendida', fontsize=10)
        plt.ylabel('Calificación Promedio', fontsize=10)
        plt.colorbar(scatter, label='Cantidad Vendida')
        plt.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

# 4. Resumen global: Top 10 productos más vendidos en todas las tiendas
ventas_globales = todas_tiendas[producto_col].value_counts().reset_index()
ventas_globales.columns = ['Producto', 'Cantidad']
top10_global = ventas_globales.head(10)

plt.figure(figsize=(12, 8))
bars = plt.barh(top10_global['Producto'], top10_global['Cantidad'],
                color='skyblue', edgecolor='navy', alpha=0.7)

for bar in bars:
    width = bar.get_width()
    plt.text(width + 0.5, bar.get_y() + bar.get_height()/2,
             f'{int(width)}', va='center')

plt.title('Top 10 Productos Más Vendidos en Todas las Tiendas', fontsize=16)
plt.xlabel('Cantidad Vendida', fontsize=14)
plt.ylabel('Producto', fontsize=14)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# 5. Distribución de ventas por tipo de producto
categoria_col = 'Categoría del Producto'
if categoria_col not in tienda1.columns:
    posibles_columnas = [col for col in tienda1.columns if 'categ' in col.lower()]
    if posibles_columnas:
        categoria_col = posibles_columnas[0]

if categoria_col in todas_tiendas.columns:
    plt.figure(figsize=(16, 12))

    for i, nombre_tienda in enumerate(nombres_tiendas, 1):
        plt.subplot(2, 2, i)

        df = tiendas_df[i-1]

        categoria_producto = df.groupby([categoria_col, producto_col]).size().reset_index()
        categoria_producto.columns = ['Categoría', 'Producto', 'Cantidad']

        top_por_categoria = categoria_producto.loc[categoria_producto.groupby('Categoría')['Cantidad'].idxmax()]

        top_categorias = top_por_categoria.nlargest(5, 'Cantidad')

        bars = plt.bar(top_categorias['Categoría'], top_categorias['Cantidad'],
                      color='lightblue', edgecolor='navy', alpha=0.7)

        for j, bar in enumerate(bars):
            height = bar.get_height()
            producto = top_categorias.iloc[j]['Producto']
            cantidad = top_categorias.iloc[j]['Cantidad']

            plt.text(bar.get_x() + bar.get_width()/2, height + 0.3,
                    f'{int(cantidad)}', ha='center')

            plt.text(bar.get_x() + bar.get_width()/2, height/2,
                    f'{producto}', ha='center', rotation=90,
                    fontsize=8, fontweight='bold', color='darkblue')

        plt.title(f'Producto Más Vendido por Categoría - {nombre_tienda}', fontsize=14)
        plt.xlabel('Categoría', fontsize=12)
        plt.ylabel('Cantidad Vendida', fontsize=12)
        plt.xticks(rotation=45, ha='right')
        plt.grid(axis='y', linestyle='--', alpha=0.7)

    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)


tienda1['tienda'] = 'Tienda 1'
tienda2['tienda'] = 'Tienda 2'
tienda3['tienda'] = 'Tienda 3'
tienda4['tienda'] = 'Tienda 4'


def estandarizar_columnas(df):

    mapeo = {
        'precio': 'Precio',
        'producto': 'Producto',
        'categoria del producto': 'Categoria del Producto',
        'categoría del producto': 'Categoria del Producto',
        'costo de envío': 'Costo de envío',
        'costo de envio': 'Costo de envío',
        'fecha de compra': 'Fecha de Compra',
        'vendedor': 'Vendedor',
        'lugar de compra': 'Lugar de Compra',
        'calificación': 'Calificación',
        'calificacion': 'Calificación',
        'método de pago': 'Método de pago',
        'metodo de pago': 'Método de pago',
        'cantidad': 'Cantidad'
    }


    renombrar = {}
    for col in df.columns:
        col_lower = col.lower()
        if col_lower in mapeo and col != mapeo[col_lower]:
            renombrar[col] = mapeo[col_lower]

    # Renombrar columnas si es necesario
    if renombrar:
        df = df.rename(columns=renombrar)

    return df


# Combinar datos de todas las tiendas
todas_tiendas = pd.concat([tienda1, tienda2, tienda3, tienda4])

print("\nPrimeras filas de los datos combinados:")
print(todas_tiendas.head())

# Análisis 1: Ingresos totales por tienda
print("\nVerificando columnas 'Precio' y 'Cantidad':")
print(f"'Precio' en columnas: {'Precio' in todas_tiendas.columns}")
print(f"'Cantidad' en columnas: {'Cantidad' in todas_tiendas.columns}")


if 'Precio' in todas_tiendas.columns and 'Cantidad' in todas_tiendas.columns:
    todas_tiendas['ingreso'] = todas_tiendas['Precio'] * todas_tiendas['Cantidad']
    print("\nColumna 'ingreso' creada correctamente")
else:
    print("\nERROR: No se encontraron las columnas 'Precio' y/o 'Cantidad'")
    # Intentar encontrar las columnas correctas
    precio_cols = [col for col in todas_tiendas.columns if 'precio' in col.lower()]
    cantidad_cols = [col for col in todas_tiendas.columns if 'cantidad' in col.lower()]
    print(f"Posibles columnas de precio: {precio_cols}")
    print(f"Posibles columnas de cantidad: {cantidad_cols}")

    # Si encontramos columnas alternativas, las usamos
    if precio_cols and cantidad_cols:
        todas_tiendas['ingreso'] = todas_tiendas[precio_cols[0]] * todas_tiendas[cantidad_cols[0]]
        print(f"Columna 'ingreso' creada usando {precio_cols[0]} y {cantidad_cols[0]}")

# Verificar que la columna ingreso existe
if 'ingreso' in todas_tiendas.columns:
    # Agrupar por tienda y sumar ingresos
    ingresos_tienda = todas_tiendas.groupby('tienda')['ingreso'].sum().reset_index()
    print("\nIngresos totales por tienda:")
    print(ingresos_tienda)
else:
    print("\nERROR: No se pudo crear la columna 'ingreso'")
    # Alternativa: mostrar la suma del precio por tienda
    if 'Precio' in todas_tiendas.columns:
        suma_precios = todas_tiendas.groupby('tienda')['Precio'].sum().reset_index()
        print("Suma de precios por tienda (no es ingreso real sin cantidad):")
        print(suma_precios)

# Visualización de ingresos por tienda
if 'ingreso' in todas_tiendas.columns:
    plt.figure(figsize=(10, 6))
    sns.barplot(x='tienda', y='ingreso', data=ingresos_tienda)
    plt.title('Ingresos Totales por Tienda')
    plt.xlabel('Tienda')
    plt.ylabel('Ingresos Totales')
    plt.tight_layout()
    plt.show()

# Análisis 2: Categorías más vendidas por tienda
categoria_cols = [col for col in todas_tiendas.columns if 'categ' in col.lower()]
categoria_col = None

if categoria_cols:
    categoria_col = categoria_cols[0]
    print(f"\nUsando columna de categoría: {categoria_col}")
else:
    print("\nERROR: No se encontró columna de categoría")

if categoria_col:
    # Conteo de ventas por categoría y tienda
    categorias_vendidas = todas_tiendas.groupby(['tienda', categoria_col]).size().reset_index(name='conteo')

    # Top 3 categorías más vendidas por tienda
    top_categorias = categorias_vendidas.sort_values(['tienda', 'conteo'], ascending=[True, False])

    print("\nCategorías más vendidas por tienda:")
    for tienda in top_categorias['tienda'].unique():
        print(f"\n{tienda}:")
        top = top_categorias[top_categorias['tienda'] == tienda].head(3)
        print(top[[categoria_col, 'conteo']])

    # Visualización de categorías más vendidas (top 3) por tienda
    plt.figure(figsize=(14, 8))
    for i, tienda in enumerate(top_categorias['tienda'].unique(), 1):
        plt.subplot(2, 2, i)
        data = top_categorias[top_categorias['tienda'] == tienda].sort_values('conteo', ascending=False).head(5)
        sns.barplot(x='conteo', y=categoria_col, data=data)
        plt.title(f'Top 5 Categorías - {tienda}')
        plt.tight_layout()
    plt.show()

# Análisis 3: Calificaciones promedio por tienda
calificacion_cols = [col for col in todas_tiendas.columns if 'calif' in col.lower()]
calificacion_col = None

if calificacion_cols:
    calificacion_col = calificacion_cols[0]
    print(f"\nUsando columna de calificación: {calificacion_col}")
else:
    print("\nERROR: No se encontró columna de calificación")

if calificacion_col:
    calificacion_promedio = todas_tiendas.groupby('tienda')[calificacion_col].mean().reset_index()
    print("\nCalificación promedio por tienda:")
    print(calificacion_promedio)


    plt.figure(figsize=(10, 6))
    sns.barplot(x='tienda', y=calificacion_col, data=calificacion_promedio)
    plt.title('Calificación Promedio por Tienda')
    plt.xlabel('Tienda')
    plt.ylabel('Calificación Promedio')
    plt.ylim(0, 5)  # Ajustar escala si es necesario
    plt.tight_layout()
    plt.show()

# Análisis 4: Productos más y menos vendidos
producto_cols = [col for col in todas_tiendas.columns if 'produc' in col.lower()]
producto_col = None

if producto_cols:
    producto_col = producto_cols[0]
    print(f"\nUsando columna de producto: {producto_col}")
else:
    print("\nERROR: No se encontró columna de producto")

if producto_col:

    productos_vendidos = todas_tiendas.groupby(['tienda', producto_col]).size().reset_index(name='conteo')

    print("\nProductos más vendidos por tienda:")
    for tienda in productos_vendidos['tienda'].unique():
        print(f"\n{tienda}:")
        top = productos_vendidos[productos_vendidos['tienda'] == tienda].sort_values('conteo', ascending=False).head(3)
        print(top[[producto_col, 'conteo']])

    print("\nProductos menos vendidos por tienda:")
    for tienda in productos_vendidos['tienda'].unique():
        print(f"\n{tienda}:")
        bottom = productos_vendidos[productos_vendidos['tienda'] == tienda].sort_values('conteo').head(3)
        print(bottom[[producto_col, 'conteo']])

# Análisis 5: Coste de envío promedio por tienda
costo_envio_cols = [col for col in todas_tiendas.columns if 'costo' in col.lower() and 'env' in col.lower()]
costo_envio_col = None

if costo_envio_cols:
    costo_envio_col = costo_envio_cols[0]
    print(f"\nUsando columna de costo de envío: {costo_envio_col}")
else:
    print("\nERROR: No se encontró columna de costo de envío")

if costo_envio_col:
    # Costo de envío promedio por tienda
    costo_envio_promedio = todas_tiendas.groupby('tienda')[costo_envio_col].mean().reset_index()
    print("\nCosto de envío promedio por tienda:")
    print(costo_envio_promedio)

    # Visualización de costo de envío promedio
    plt.figure(figsize=(10, 6))
    sns.barplot(x='tienda', y=costo_envio_col, data=costo_envio_promedio)
    plt.title('Costo de Envío Promedio por Tienda')
    plt.xlabel('Tienda')
    plt.ylabel('Costo de Envío Promedio')
    plt.tight_layout()
    plt.show()

# Análisis adicional: Rentabilidad (Ingresos - Costos de envío)
if 'ingreso' in todas_tiendas.columns and costo_envio_col:
    print("\nCalculando rentabilidad...")
    # Asegurarse de que ambas columnas sean numéricas
    todas_tiendas['ingreso'] = pd.to_numeric(todas_tiendas['ingreso'], errors='coerce')
    todas_tiendas[costo_envio_col] = pd.to_numeric(todas_tiendas[costo_envio_col], errors='coerce')

    # Calcular rentabilidad
    todas_tiendas['rentabilidad'] = todas_tiendas['ingreso'] - todas_tiendas[costo_envio_col]

    # Agrupar por tienda
    rentabilidad_tienda = todas_tiendas.groupby('tienda')['rentabilidad'].sum().reset_index()
    print("\nRentabilidad total por tienda (Ingresos - Costos de envío):")
    print(rentabilidad_tienda)

    # Visualización de rentabilidad
    plt.figure(figsize=(10, 6))
    sns.barplot(x='tienda', y='rentabilidad', data=rentabilidad_tienda)
    plt.title('Rentabilidad Total por Tienda')
    plt.xlabel('Tienda')
    plt.ylabel('Rentabilidad (Ingresos - Costos de envío)')
    plt.tight_layout()
    plt.show()
else:
    print("\nNo se puede calcular rentabilidad: faltan columnas necesarias")

# Análisis 6: Método de pago más común por tienda
metodo_pago_cols = [col for col in todas_tiendas.columns if 'pago' in col.lower()]
metodo_pago_col = None

if metodo_pago_cols:
    metodo_pago_col = metodo_pago_cols[0]
    print(f"\nUsando columna de método de pago: {metodo_pago_col}")
else:
    print("\nERROR: No se encontró columna de método de pago")

if metodo_pago_col:
    # Conteo de métodos de pago por tienda
    metodos_pago = todas_tiendas.groupby(['tienda', metodo_pago_col]).size().reset_index(name='conteo')
    print("\nMétodos de pago más comunes por tienda:")
    for tienda in metodos_pago['tienda'].unique():
        print(f"\n{tienda}:")
        top = metodos_pago[metodos_pago['tienda'] == tienda].sort_values('conteo', ascending=False)
        print(top[[metodo_pago_col, 'conteo']])
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import LinearSegmentedColormap
import warnings
warnings.filterwarnings('ignore')


plt.style.use('seaborn-v0_8')
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 12


url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"


print("Cargando datos de las 4 tiendas...")
tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)


tienda1['Tienda'] = 'Tienda 1'
tienda2['Tienda'] = 'Tienda 2'
tienda3['Tienda'] = 'Tienda 3'
tienda4['Tienda'] = 'Tienda 4'

# Combinar todos los datos
df_completo = pd.concat([tienda1, tienda2, tienda3, tienda4], ignore_index=True)

print(f"Datos cargados exitosamente:")
print(f"- Total de registros: {len(df_completo)}")
print(f"- Tienda 1: {len(tienda1)} registros")
print(f"- Tienda 2: {len(tienda2)} registros")
print(f"- Tienda 3: {len(tienda3)} registros")
print(f"- Tienda 4: {len(tienda4)} registros")

# Verificar columnas disponibles
print(f"\nColumnas disponibles: {list(df_completo.columns)}")

# Verificar datos geográficos
print(f"\nAnálisis inicial de coordenadas:")
print(f"Rango de latitud: {df_completo['lat'].min():.4f} a {df_completo['lat'].max():.4f}")
print(f"Rango de longitud: {df_completo['lon'].min():.4f} a {df_completo['lon'].max():.4f}")

# Calcular métricas por tienda
print("\n" + "="*60)
print("RESUMEN POR TIENDA")
print("="*60)

resumen_tiendas = df_completo.groupby('Tienda').agg({
    'Precio': ['mean', 'sum', 'count'],
    'Costo de envío': 'mean',
    'Calificación': 'mean',
    'lat': ['mean', 'std'],
    'lon': ['mean', 'std']
}).round(2)

resumen_tiendas.columns = ['Precio_Promedio', 'Ingresos_Total', 'Num_Ventas',
                          'Costo_Envio_Promedio', 'Calificacion_Promedio',
                          'Lat_Promedio', 'Lat_Std', 'Lon_Promedio', 'Lon_Std']

print(resumen_tiendas)

# 1. MAPA DE DISPERSIÓN DE VENTAS POR UBICACIÓN
print("\n" + "="*60)
print("1. GENERANDO MAPA DE DISPERSIÓN DE VENTAS")
print("="*60)

fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Distribución Geográfica de Ventas por Tienda', fontsize=16, fontweight='bold')

colores = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4']
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

for i, (tienda, color) in enumerate(zip(tiendas, colores)):
    row = i // 2
    col = i % 2

    datos_tienda = df_completo[df_completo['Tienda'] == tienda]

    scatter = axes[row, col].scatter(datos_tienda['lon'], datos_tienda['lat'],
                                   c=datos_tienda['Precio'], cmap='viridis',
                                   alpha=0.7, s=60, edgecolors='black', linewidth=0.5)

    axes[row, col].set_title(f'{tienda} - {len(datos_tienda)} ventas', fontweight='bold')
    axes[row, col].set_xlabel('Longitud')
    axes[row, col].set_ylabel('Latitud')
    axes[row, col].grid(True, alpha=0.3)

    # Añadir colorbar para cada subplot
    cbar = plt.colorbar(scatter, ax=axes[row, col])
    cbar.set_label('Precio ($)', rotation=270, labelpad=15)

plt.tight_layout()
plt.show()

# 2. MAPA GENERAL DE TODAS LAS TIENDAS
print("\n2. GENERANDO MAPA GENERAL DE TODAS LAS TIENDAS")

plt.figure(figsize=(14, 10))
for i, (tienda, color) in enumerate(zip(tiendas, colores)):
    datos_tienda = df_completo[df_completo['Tienda'] == tienda]
    plt.scatter(datos_tienda['lon'], datos_tienda['lat'],
               c=color, label=tienda, alpha=0.7, s=50, edgecolors='black', linewidth=0.5)

plt.title('Distribución Geográfica de Todas las Tiendas', fontsize=16, fontweight='bold')
plt.xlabel('Longitud', fontsize=12)
plt.ylabel('Latitud', fontsize=12)
plt.legend(title='Tienda', fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# 3. MAPA DE CALOR DE DENSIDAD DE VENTAS
print("\n3. GENERANDO MAPA DE CALOR DE DENSIDAD")

plt.figure(figsize=(14, 10))

# Crear bins para el heatmap
lat_bins = np.linspace(df_completo['lat'].min(), df_completo['lat'].max(), 20)
lon_bins = np.linspace(df_completo['lon'].min(), df_completo['lon'].max(), 20)

# Crear matriz de densidad
hist, lat_edges, lon_edges = np.histogram2d(df_completo['lat'], df_completo['lon'],
                                           bins=[lat_bins, lon_bins])

# Crear el heatmap
plt.imshow(hist, extent=[lon_edges[0], lon_edges[-1], lat_edges[0], lat_edges[-1]],
          cmap='YlOrRd', aspect='auto', origin='lower', alpha=0.8)

plt.colorbar(label='Número de Ventas')
plt.title('Mapa de Calor - Densidad de Ventas', fontsize=16, fontweight='bold')
plt.xlabel('Longitud', fontsize=12)
plt.ylabel('Latitud', fontsize=12)

# Superponer puntos de tiendas
for i, (tienda, color) in enumerate(zip(tiendas, colores)):
    datos_tienda = df_completo[df_completo['Tienda'] == tienda]
    plt.scatter(datos_tienda['lon'], datos_tienda['lat'],
               c=color, label=tienda, alpha=0.6, s=30, edgecolors='white', linewidth=0.5)

plt.legend(title='Tienda', loc='upper right')
plt.tight_layout()
plt.show()

# 4. ANÁLISIS DE INGRESOS POR REGIÓN
print("\n" + "="*60)
print("4. ANÁLISIS DE INGRESOS POR REGIÓN")
print("="*60)

# Dividir el área en cuadrantes
lat_media = df_completo['lat'].median()
lon_media = df_completo['lon'].median()

def asignar_cuadrante(row):
    if row['lat'] >= lat_media and row['lon'] >= lon_media:
        return 'Noreste'
    elif row['lat'] >= lat_media and row['lon'] < lon_media:
        return 'Noroeste'
    elif row['lat'] < lat_media and row['lon'] >= lon_media:
        return 'Sureste'
    else:
        return 'Suroeste'

df_completo['Cuadrante'] = df_completo.apply(asignar_cuadrante, axis=1)

# Análisis por cuadrante
analisis_cuadrante = df_completo.groupby(['Cuadrante', 'Tienda']).agg({
    'Precio': ['sum', 'mean', 'count'],
    'Calificación': 'mean'
}).round(2)

analisis_cuadrante.columns = ['Ingresos_Total', 'Precio_Promedio', 'Num_Ventas', 'Calificacion_Promedio']
print(analisis_cuadrante)

# Visualización de ingresos por cuadrante
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Gráfico de ingresos por cuadrante
ingresos_cuadrante = df_completo.groupby('Cuadrante')['Precio'].sum().sort_values(ascending=False)
axes[0].bar(ingresos_cuadrante.index, ingresos_cuadrante.values,
           color=['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4'])
axes[0].set_title('Ingresos Totales por Cuadrante', fontweight='bold')
axes[0].set_ylabel('Ingresos ($)')
axes[0].tick_params(axis='x', rotation=45)

# Gráfico de número de ventas por cuadrante
ventas_cuadrante = df_completo.groupby('Cuadrante').size().sort_values(ascending=False)
axes[1].bar(ventas_cuadrante.index, ventas_cuadrante.values,
           color=['#FFB74D', '#81C784', '#64B5F6', '#F06292'])
axes[1].set_title('Número de Ventas por Cuadrante', fontweight='bold')
axes[1].set_ylabel('Número de Ventas')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# 5. ANÁLISIS DE CALIFICACIONES POR UBICACIÓN
print("\n5. ANÁLISIS DE CALIFICACIONES POR UBICACIÓN")

plt.figure(figsize=(12, 8))
scatter = plt.scatter(df_completo['lon'], df_completo['lat'],
                     c=df_completo['Calificación'], cmap='RdYlGn',
                     s=60, alpha=0.7, edgecolors='black', linewidth=0.5)

plt.colorbar(scatter, label='Calificación (1-5)')
plt.title('Mapa de Calificaciones por Ubicación', fontsize=16, fontweight='bold')
plt.xlabel('Longitud', fontsize=12)
plt.ylabel('Latitud', fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# 6. ANÁLISIS ESTADÍSTICO DE PATRONES GEOGRÁFICOS
print("\n" + "="*60)
print("6. ANÁLISIS ESTADÍSTICO DE PATRONES")
print("="*60)

# Correlación entre ubicación y variables de negocio
correlaciones = df_completo[['lat', 'lon', 'Precio', 'Costo de envío', 'Calificación']].corr()
print("Matriz de correlación:")
print(correlaciones.round(3))

# Análisis de dispersión geográfica por tienda
print(f"\nDispersión geográfica por tienda:")
for tienda in tiendas:
    datos_tienda = df_completo[df_completo['Tienda'] == tienda]
    lat_std = datos_tienda['lat'].std()
    lon_std = datos_tienda['lon'].std()
    area_cobertura = (datos_tienda['lat'].max() - datos_tienda['lat'].min()) * \
                    (datos_tienda['lon'].max() - datos_tienda['lon'].min())

    print(f"{tienda}:")
    print(f"  - Desviación estándar latitud: {lat_std:.4f}")
    print(f"  - Desviación estándar longitud: {lon_std:.4f}")
    print(f"  - Área de cobertura aproximada: {area_cobertura:.4f} grados²")

# 7. RENDIMIENTO POR DISTANCIA AL CENTRO
print("\n7. ANÁLISIS DE RENDIMIENTO POR DISTANCIA AL CENTRO")

# Calcular centro geográfico
centro_lat = df_completo['lat'].mean()
centro_lon = df_completo['lon'].mean()

# Calcular distancia al centro (aproximada)
df_completo['distancia_centro'] = np.sqrt((df_completo['lat'] - centro_lat)**2 +
                                        (df_completo['lon'] - centro_lon)**2)

# Dividir en zonas por distancia
df_completo['zona_distancia'] = pd.cut(df_completo['distancia_centro'],
                                      bins=3, labels=['Cerca', 'Media', 'Lejos'])

# Análisis por zona de distancia
analisis_distancia = df_completo.groupby('zona_distancia').agg({
    'Precio': ['mean', 'sum'],
    'Calificación': 'mean',
    'Costo de envío': 'mean'
}).round(2)

analisis_distancia.columns = ['Precio_Promedio', 'Ingresos_Total', 'Calificacion_Promedio', 'Costo_Envio_Promedio']
print("\nRendimiento por zona de distancia al centro:")
print(analisis_distancia)

# Gráfico de rendimiento por distancia
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Precio promedio por zona
axes[0].bar(analisis_distancia.index, analisis_distancia['Precio_Promedio'],
           color=['#66BB6A', '#42A5F5', '#EF5350'])
axes[0].set_title('Precio Promedio por Zona de Distancia', fontweight='bold')
axes[0].set_ylabel('Precio Promedio ($)')

# Calificación promedio por zona
axes[1].bar(analisis_distancia.index, analisis_distancia['Calificacion_Promedio'],
           color=['#AB47BC', '#FF7043', '#26A69A'])
axes[1].set_title('Calificación Promedio por Zona de Distancia', fontweight='bold')
axes[1].set_ylabel('Calificación Promedio')

# Costo de envío por zona
axes[2].bar(analisis_distancia.index, analisis_distancia['Costo_Envio_Promedio'],
           color=['#FFCA28', '#5C6BC0', '#8D6E63'])
axes[2].set_title('Costo de Envío Promedio por Zona de Distancia', fontweight='bold')
axes[2].set_ylabel('Costo de Envío Promedio ($)')

plt.tight_layout()
plt.show()

# 8. CONCLUSIONES Y RECOMENDACIONES
print("\n" + "="*60)
print("8. CONCLUSIONES Y RECOMENDACIONES")
print("="*60)

# Tienda con mejor rendimiento geográfico
mejor_tienda = resumen_tiendas['Ingresos_Total'].idxmax()
peor_tienda = resumen_tiendas['Ingresos_Total'].idxmin()

print(f"CONCLUSIONES PRINCIPALES:")
print(f"")
print(f"1. RENDIMIENTO POR TIENDA:")
print(f"   - {mejor_tienda} tiene los mayores ingresos totales")
print(f"   - {peor_tienda} tiene los menores ingresos totales")
print(f"")

# Cuadrante más rentable
cuadrante_mas_rentable = ingresos_cuadrante.index[0]
print(f"2. DISTRIBUCIÓN GEOGRÁFICA:")
print(f"   - El cuadrante {cuadrante_mas_rentable} es el más rentable")
print(f"   - Concentración de ventas: {ingresos_cuadrante.iloc[0]:,.2f} en ingresos")
print(f"")

# Correlaciones importantes
corr_precio_lat = correlaciones.loc['Precio', 'lat']
corr_precio_lon = correlaciones.loc['Precio', 'lon']
print(f"3. PATRONES GEOGRÁFICOS:")
print(f"   - Correlación Precio-Latitud: {corr_precio_lat:.3f}")
print(f"   - Correlación Precio-Longitud: {corr_precio_lon:.3f}")
print(f"")

print(f"4. RECOMENDACIONES:")
print(f"   - Enfocar esfuerzos de marketing en el cuadrante {cuadrante_mas_rentable}")
print(f"   - Analizar por qué {mejor_tienda} tiene mejor rendimiento")
print(f"   - Considerar establecer centros de distribución en áreas de alta densidad")
print(f"   - Optimizar costos de envío según la distancia al centro")
```
